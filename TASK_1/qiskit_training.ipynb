```python
# LEARNING THE "NOT" GATE 

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.x(0) # I am "notting" the first qubit, so now it should be 1 since it is initialized as 0

pdc(qc)

sv = Statevector(qc)
print("State after X gate:")
print(sv)

# the output is where:
# index 0 is the probabilty of it being q1 = 0 and q0 = 0
# index 1 is the probabiity of it being q1 = 0 and q0 = 1
# index 2 is the probabilty of it being q1 = 1 and q0 = 0
# index 3 is the probabilty of it being q1 = 1 and qo = 1
```

```python
# LEARNING THE "h()" HADAMARD GATE
# FMI: [1, 0] is a column vector, meaning |0>, while [0, 1] means |1>

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0) # now the qubit is in superposition, meaning 50/50 chance of it being either "1" or "0" until measured

pdc

sv = Statevector(qc)
print("State after H gate:")
print(sv)

# What happens if you apply "h()" function twice
qc2 = QuantumCircuit(1)
qc2.h(0)
qc2.h(0) # H is its own inverse, so it should go back to "0"

pdc(qc)

sv2 = Statevector(qc2)
print("State after double H gate")
print(sv2)
```

```python
# LEARNING THE "z()" GATE, WHICH FLIPS THE PHASE (OR SIGN) OF A QUBIT WITHOUT CHANGING PROBABILITY

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0) # must be in superposition first
qc.z(0) # flip the phase of |1> component

pdc(qc)

sv = Statevector(qc)
print("State after Z gate")
print(sv)
```

```python
# LEARNING THE "y()" GATE, WHICH ROTATES THE QUBIT AROUND THE Y-AXIS OF BLOCH SPHERE

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.y(0)

pdc

sv = Statevector(qc)
print("State after Y gate:")
print(sv)
```

```python
# LEARNING THE "s()" GATE, WHICH ROTATES THE QUBIT PHASE BY 90 DEGREES, LIKE DOING HALF OF "z()" GATE

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0) # into superposition first
qc.s(0) # Add 90 degree phase to |1> component

pdc

sv = Statevector(qc)
print("State after Y gate:")
print(sv)

# S dagger (inverse of S gate) = sdg
qc2 = QuantumCircuit(1)
qc2.h(0)
qc2.sdg(0) # negative 90 degree phase to |1> component
```

```python
# LEARNING THE "t()" GATE, WHICH ROTATES THE QUBIT PHASE BY 45 DEGREES, LIKE DOING HALF OF "s()" GATE

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0) # first superposition
qc.t(0) # rotate phase by 45 degree

pdc(qc)

sv = Statevector(qc)
print("State after T gate:")
print(sv)

# T dagger (inverse of normal T gate)
qc2 = QuantumCircuit(1)
qc2.tdg(0) # rotate phase by negative 45 degree
```

```python
# LEARNING "rx()", "ry()", and "rz()" FUNCTION TO ROTATE AROUND DIFFERENT AXIS

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.circuit import Parameter
import numpy as np

qc = QuantumCircuit(1)

qc.rx(np.pi/2, 0) # 90 degree rotation (pi/2)

pdc

sv = Statevector(qc)
print("After rotating 90 degrees around x-axis")
print(sv)

qc2 = QuantumCircuit(1)
qc2.ry(np.pi, 0) # 180 degree rotation (pi)

pdc(qc2)

theta = Parameter('90')
qc3 = QuantumCircuit(1)
qc3.rz(theta, 0) # rotate around Z axis by theta

pdc
```

## MULTI-QUBIT GATES

```python
# LEARNING THE "cx()" GATE, WHICH FLIPS "target" QUBIT ONLY IF THE CONTROL QUBIT IS |1>. THIS CREATES AN ENTANGLEMENT

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.h(0) # put q0 in superposition (50/50 chance of either being 1 or 0 until measured)
qc.cx(0, 1) # entangle q0 with q1

sv = Statevector(qc)
print("Bell state (entangled):")
print(sv)
# Shows: [0.707, 0, 0, 0.707]
# This means: 50% chance of |00>, 50% chance of |11>
# NEVER |01> or |10>

# Measuring one qubit instantly determines the other qubit (if you measure q0 to be N, q1 will also be N)
qc.measure_all()
pdc(qc)
```

```python
# LEARNING THE "cz()" GATE, SIMILAR TO "cx()", APPLIES A Z GATE (PHASE FLIP) TO THE TARGET WHEN CONTROL IS |1>

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.h(0)
qc.h(1) # need to put both qubits in superposition
qc.cz(0, 1) # phase flip on |11> component only

sv = Statevector(qc)
print("After putting both qubits in superposition, then a CZ gate:")
print(sv)
# [0.5, 0.5, 0.5, -0.5] - a phase flip (sign change) on |11> component

pdc(qc)
```

```python
# LEARNING THE "ccx()" GATE, WHICH IS BASICALLY A QUANTUM "AND" GATE. FLIPS "target" ONLY IF BOTH CONTROL QUBITS ARE |1>

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(3)
qc.x(0) 
qc.x(1) # set both qubits to 1 for this to flip
# now, q2 is still |0>

qc.ccx(0, 1, 2) # Toffoli: flip q2 only if q1 and q0 are |1> (which in this case they are)

pdc

sv = Statevector(qc)
print("After Toffoli (both control q0 and q1 are |1>)")
print(sv)

# output: q2 should be flipped to be |1>, since q1 and q0 are |1>

qc2 = QuantumCircuit(3)
qc2.x(0)
qc2.x(2)

qc2.ccx(0, 2, 1)

pdc(qc2)

sv = Statevector(qc2)
print("After Toffoli (both control q0 and q2 are |1>)")
print(sv)
```

```python
# LEARNING THE "swap()" METHOD, TO SWAP THE STATES OF TWO QUBITS, USED TO MOVE QUANTUM INFORMATION AROUND

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.x(0) # make q0 to be |1>
# q1 will still be |0>

sv_before = Statevector(qc)
print("Before swapping: ", sv_before)

pdc

# now lets swap the values so q0 = |0> and q1 = |1>
qc.swap(0, 1)

pdc

sv_after = Statevector(qc)
print("After swapping q0 and q1: ", sv_after)
```

```python
# LEARNING THE "ch()" METHOD, WHICH APPLYS A CONTROLLED HADAMARD GATE ONLY IF THE CONTROL BIT IS |1>

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

qc = QuantumCircuit(2)
qc.x(0) # set control to |1>

# Controlled-H: apply H to q1 only if q0 is |1>
qc.ch(0, 1)

pdc

sv = Statevector(qc)
print("After X and CH: ", sv)

qc2 = QuantumCircuit(2)
qc2.x(0)
qc2.cry(np.pi/4, 0, 1) # controlled Ry(pi/4)
pdc(qc2)
```

## RUNNING & SIMULATING CIRCUITS

```python
# USING STATEVECTOR

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

# get the full quantum state
sv = Statevector(qc)
print(sv) # raw complex amplitudes
sv.draw('text') # formatted text output
sv.draw('latex') # LaTex output

print()
print()

# get probabilties from the statevector
probs = sv.probabilities_dict()
print("Measurement probabilities:", probs)

print()
print()

# simulate a measurement
result, new_state = sv.measure()
print("Measured:", result)
print("State after measurement:", new_state)
```

```python
# USING AERSIMULATOR - RUNNING SHOTS

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2, 2) # 2 qubits, 2 bits
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

# set up the simulator
simulator = AerSimulator()

# transpile (optimize circuit for the backend)
compiled = transpile(qc, simulator)

# run 1000000 shots
result = simulator.run(compiled, shots=1000000).result()
counts = result.get_counts()

print("Results:", counts)
```

```python
# MODERN QISKIT PRIMITIVE API

from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

sampler = StatevectorSampler()
job = sampler.run([qc], shots=1000)
result = job.result()

# get counts from the result
pub_result = result[0]
counts = pub_result.data.meas.get_counts()
print("Counts: ", counts)
```

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorEstimator
from qiskit.quantum_info import SparsePauliOp

# create a circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

# define what we want to measure: <ZZ> operator
# ZZ measures correlation between the two qubits
observable = SparsePauliOp("ZZ")

estimator = StatevectorEstimator()
job = estimator.run([(qc, observable)])
result = job.result()

expectation_value = result[0].data.evs
print(f"Expectation Value of ZZ: {expectation_value}")
```

## CIRCUIT OPERATIONS AND TOOLS

```python
# LEARNING THE "barrier()" METHOD, WHICH ADDS A VISUAL LINE IN CIRCUIT DIAGRAM, GOOD FOR DEBUGGING AND COMPILING SEGMENTS WITHIN BARRIER

from qiskit import QuantumCircuit

qc = QuantumCircuit(3)

# put all bits in superposition
qc.h(0)
qc.h(1)
qc.h(2)

qc.barrier() # visual seperator

qc.cx(0, 1)
qc.cx(1, 2)

qc.barrier() # visual seperator

qc.measure_all()

pdc(qc)
```

```python
# LEARNING THE "reset()" METHOD, WHICH IS USED TO RESET A QUBIT BACK TO |0> STATE, REGARDLESS OF WHAT IT IS CURRENTLY

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.x(0)

# now the qubit is |1>, lets reset it back to |0>

sv = Statevector(qc)
print("Before reset:")
print(sv)

print()
print()

pdc(qc)

qc.reset(0)

sv = Statevector(qc)
print("After reset:")
print(sv)

pdc(qc)
```

```python
# LEARNING THE "transpile()" METHOD, WHICH OPTIMIZES CIRCUIT TO RUN ON QUANTUM HARDWARE

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 2) # on a real hardware, q0 and q2 may not be neighboring, so lets see how transpile optimizes

pdc(qc)

# now transpile for a simulator backend
backend = AerSimulator()
transpiled = transpile(qc, backend, optimization_level=3)

print("After transpiled:")
pdc(transpiled)
```

```python
# LEARNING PARAMETERIZED CIRCUITS (USEFUL FOR QUANTUM MACHINE LEARNING)

from qiskit import QuantumCircuit
from qiskit.circuit import Parameter, ParameterVector
import numpy as np

theta = Parameter('0')
qc = QuantumCircuit(1)
qc.ry(theta, 0)

bound = qc.assign_parameters({theta: np.pi/4})
print("Bound Circuit: ")
pdc(qc)

params = ParameterVector('x', 4)
qc2 = QuantumCircuit(2)
qc2.ry(params[0], 0)
qc2.ry(params[1], 1)
qc2.cx(0, 1)
qc2.rz(params[2], 0)
qc2.rz(params[3], 1)

bound2 = qc2.assign_parameters([0.1, 0.2, 0.3, 0.4])
pdc(qc2)
```

```python
# LEARNING THE "compose()" METHOD, TO COMBINE DIFFERENT CIRCUITS

from qiskit import QuantumCircuit

first_circuit = QuantumCircuit(2)
first_circuit.h(0)
first_circuit.cx(0, 1)

pdc(first_circuit)

second_circuit = QuantumCircuit(2)
second_circuit.h(0)
second_circuit.z(1)

pdc(second_circuit)

combined_circuit = first_circuit.compose(second_circuit)

pdc(combined_circuit)


# now lets compose with specific qubit mapping
full2 = QuantumCircuit(4)
full2.compose(first_circuit, qubits=[0, 1], inplace=True)
full2.compose(second_circuit, qubits=[0, 1], inplace=True)

pdc(full2)
```

```python
# LEARNING THE ".inverse()" METHOD TO GET THE INVERSE OF A CIRCUIT, AND SEE HOW IT TURNS TO IDENTITY WHEN COMBINED WITH ORIGINAL CIRCUIT

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.t(0)

pdc(qc)

# inverse circuit
qc_inv = qc.inverse()

pdc(qc_inv)

# combining qc with its inverse => identify circuit
combined = qc.compose(qc_inv)
sv = Statevector(combined)
print(sv)
```
