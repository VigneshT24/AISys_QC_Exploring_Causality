# **FIVE QUANTUM CIRCUIT USED FOR THIS PROJECT**

## Circuit 1: Bell State Circuit

```python
import matplotlib.pyplot as plt
from basic_quantum_circuits import BasicQuantumCircuits as BQC
import numpy as np
%matplotlib inline

def show_circuit(circuit_name, qc):
    """Prints the circuit in color using matplotlib's pyplot"""
    print(f"{circuit_name}: ")
    qc.draw('mpl')
    plt.show()

show_circuit("Bell State Circuit", BQC().bell_state_circuit())
```

## Circuit 2: GHZ State Circuit

```python
show_circuit("GHZ State Circuit", BQC().ghz_state_circuit(5))
```

## Circuit 3: Parameterized Circuit

```python
show_circuit("Parameterized Circuit", BQC().parameterized_circuit(3, [('rx', np.pi/2), ('ry', np.pi/4), ('rz', np.pi/8)]))
```

## Circuit 4: Grover Circuit

```python
show_circuit("Grover Circuit", BQC().grover_circuit(4, "1001"))
```

## Circuit 5: Variable Depth Circuit

```python
show_circuit("Variable Depth Circuit", BQC().variable_depth_circuit(4, 5))
```
